```python
import tkinter as tk
import time
from datetime import timedelta
from PIL import Image, ImageTk

class Point:
    def __init__(self, canvas, number):
        # Initial position of the point
        self.point_x, self.point_y = -20, 432
        self.number = number
        # Draw the initial point
        self.point = canvas.create_oval(self.point_x, self.point_y, self.point_x+16, self.point_y+16, fill='red')
        # Draw text on the canvas
        self.text = canvas.create_text(self.point_x+10, self.point_y+10, text=number, anchor="center", font=("Arial", 8), fill="white")
    
    def move_point(self, master, canvas, x, y, stop_x, stop_y):
        # Update the position of the point
        self.point_x += x  # You can adjust the speed by changing this value
        self.point_y += y  # Adjust as needed
        canvas.itemconfig(self.point, fill="red")

       # Complete stop condition
        x_done = (x > 0 and self.point_x >= stop_x) or (x < 0 and self.point_x <= stop_x) or x == 0
        y_done = (y > 0 and self.point_y >= stop_y) or (y < 0 and self.point_y <= stop_y) or y == 0

        if x_done and y_done:
            canvas.coords(self.point, self.point_x, self.point_y, self.point_x + 20, self.point_y + 20)
            canvas.coords(self.text, self.point_x + 10, self.point_y + 10)
            return

        # If the vertical destination is reached, change direction
        if y != 0 and ((y > 0 and self.point_y >= stop_y) or (y < 0 and self.point_y <= stop_y)):
            x = 15 if stop_x > self.point_x else -15
            y = 0
            self.point_y = stop_y   
        # Update the position of the point and text on the canvas
        canvas.coords(self.point, self.point_x, self.point_y, self.point_x+18, self.point_y+18)
        canvas.coords(self.text, self.point_x+10, self.point_y+10)
        master.after(25, lambda: self.move_point(master, canvas,x, y, stop_x, stop_y))
       
    def out_point(self, master, canvas, x, y):
        # Update the position of the point
        self.point_x += x 
        self.point_y += y  
        canvas.itemconfig(self.point, fill="green")
        # Update the position of the point and text on the canvas
        canvas.coords(self.point, self.point_x, self.point_y, self.point_x+20, self.point_y+20)
        canvas.coords(self.text, self.point_x+10, self.point_y+10)
        master.after(50, lambda: self.out_point(master, canvas, x, y))


class Queue:
    def __init__(self, master):
        self.master = master
        self.master.title("Terminal du port")
        self.canvas = tk.Canvas(self.master, width=1000, height=700, bg='white')
        self.canvas.pack()
        self.queue = []
        
        # ---------- FIRST STEP: Initialization ----------
        # Initialize
        self.chronometer_value = 0 # start timer at 0 seconds
        self.x1 = Point(self.canvas, 1) # create point object on canvas (ID = 1)
        
       # -----Our company-----
        self.images = []
        self.draw_company()

        # -----Buttons-----
        button_x1, button_y1 = 450, 630
        button_x2, button_y2 = 550, 665
        self.button_arret = tk.Button(self.master, text="Stop", command=self.stop_timer, state=tk.DISABLED)
        self.canvas.create_window((button_x1 + button_x2) / 2, (button_y1 + button_y2) / 2, window=self.button_arret,
                                 width=button_x2 - button_x1, height=button_y2 - button_y1)
        
        button_x1, button_y1 = 450, 590
        button_x2, button_y2 = 550, 620
        self.label = tk.Label(master, text="Time: 0 seconds")
        self.canvas.create_window((button_x1 + button_x2) / 2, (button_y1 + button_y2) / 2, window=self.label,
                                 width=button_x2 - button_x1, height=button_y2 - button_y1)
        
        button_x1, button_y1 = 330, 630
        button_x2, button_y2 = 430, 665
        self.button_marche = tk.Button(self.master, text="Start", command=self.start_timer)
        self.canvas.create_window((button_x1 + button_x2) / 2, (button_y1 + button_y2) / 2, window=self.button_marche,
                                 width=button_x2 - button_x1, height=button_y2 - button_y1)
        
        button_x1, button_y1 = 570, 630
        button_x2, button_y2 = 670, 665
        self.button_raz = tk.Button(self.master, text="Reset", command=self.restart_timer, state=tk.DISABLED)
        self.canvas.create_window((button_x1 + button_x2) / 2, (button_y1 + button_y2) / 2, window=self.button_raz,
                                 width=button_x2 - button_x1, height=button_y2 - button_y1)
        
        # ---------- SECOND STEP: Set up and run the timer ----------
        # Set up the timer
        self.timer_running = False # flag to control if timer is active
        self.update_timer() # start the timer update loop

    def update_timer(self):
        if self.timer_running: # only run if timer is active
            self.chronometer_value += 1 # increase time by 1 second
            self.label.config(text=f"Time: {self.chronometer_value} seconds") # update time label
             # ---- Movements based on time ----
            if self.chronometer_value == 4:
                self.x1.move_point(self.master, self.canvas, 15, 0,165, 432) # move point to position A
            if self.chronometer_value == 30:
                self.x1.move_point(self.master, self.canvas, 15, 0, 230, 432) # move to position B
            if self.chronometer_value == 31:
                self.x1.move_point(self.master, self.canvas, 15, 30, 765, 520) # move to position C  
            if self.chronometer_value == 33:
                self.x1.move_point(self.master, self.canvas, -1, -85, 270, 432) # move to position D
            if self.chronometer_value == 35:
                self.x1.move_point(self.master, self.canvas, 4, -20, 240, 370)  # move to position E
            if self.chronometer_value == 102:
                self.x1.move_point(self.master, self.canvas, 5, -30, 250, 82)  # move to position F    
            if self.chronometer_value == 103:
                self.x1.out_point(self.master, self.canvas, -30, 0) # remove point from canvas         
   
            self.master.after(1000, self.update_timer)

            

    def start_timer(self):
        self.timer_running = True
        self.button_marche.config(state=tk.DISABLED)
        self.button_arret.config(state=tk.NORMAL)
        self.button_raz.config(state=tk.NORMAL)
        self.update_timer()

    def stop_timer(self):
        self.timer_running = False
        self.button_marche.config(state=tk.NORMAL)
        self.button_arret.config(state=tk.DISABLED)
        self.button_raz.config(state=tk.NORMAL)

     # ---------- THIRD STEP: Reset / Restart Simulation ---------
    def restart_timer(self):
        self.timer_running = False
        self.button_marche.config(state=tk.NORMAL)
        self.button_raz.config(state=tk.DISABLED)
        self.button_arret.config(state=tk.DISABLED)
        # we use out_point here to delete the point from Tkinter window
        self.x1.out_point(self.master, self.canvas, 0, 2000) # delete point from window

        self.chronometer_value = 0 # reset timer value
        self.label.config(text="Time: 0 seconds") # reset label display
        
        self.x1 = Point(self.canvas, 1) # recreate point for next simulation

       
        self.update_timer()

    def draw_company(self):
        self.canvas.create_line(200,15,953,15, fill='black', width=2)
        self.canvas.create_line(200,40,900,40, fill='black', width=2)
        self.canvas.create_line(350,100,850,100, fill='black', width=2)
        self.canvas.create_line(350,110,850,110, fill='black', width=2)
        self.canvas.create_line(415,285,900,285, fill='black', width=2)
        self.canvas.create_line(440,300,951,300, fill='black', width=2)
        self.canvas.create_line(0,570,850,570, fill='black', width=4)
        self.canvas.create_line(0,550,800,550, fill='black', width=4) 
        self.canvas.create_line(370,429,800,429, fill='black', width=2) 
        self.canvas.create_line(405,401,850,401, fill='black', width=2) 
        self.canvas.create_line(799,548,799, 429, fill='black', width=2)
        self.canvas.create_line(851,572,851, 400, fill='black', width=2)
        self.canvas.create_line(950,15,950,300, fill='black', width=2)
        self.canvas.create_line(899,40,899,285, fill='black', width=2)
        self.canvas.create_line(950,15,950,300, fill='black', width=2)
        self.canvas.create_line(405,400,440, 300, fill='black', width=2)
        self.canvas.create_line(371,428,415, 285, fill='black', width=2)
        self.canvas.create_line(110,430,0, 430, fill='black', width=3)
        self.canvas.create_line(120,400,0, 400, fill='black', width=3)
        self.canvas.create_line(119,400,112,430, fill='black', width=3)
        self.canvas.create_line(350,130,850,130, fill='black', width=2)
        self.canvas.create_line(350,140,850,140, fill='black', width=2)
        
        self.canvas.create_line(350,170,850,170, fill='black', width=2)
        self.canvas.create_line(350,180,850,180, fill='black', width=2)
        
        self.canvas.create_line(350,210,850,210, fill='black', width=2)
        self.canvas.create_line(350,220,850,220, fill='black', width=2)
        
        self.canvas.create_line(350,250,850,250, fill='black', width=2)
        self.canvas.create_line(350,260,850,260, fill='black', width=2)
        #1
        self.canvas.create_line(350,145,590,145, fill='black', width=1)
        self.canvas.create_line(350,155,590,155, fill='black', width=1)
        self.canvas.create_line(350,157,590,157, fill='black', width=1)
        self.canvas.create_line(350,166,590,166, fill='black', width=1)
        
        self.canvas.create_line(600,145,850,145, fill='black', width=1)
        self.canvas.create_line(600,155,850,155, fill='black', width=1)
        self.canvas.create_line(600,157,850,157, fill='black', width=1)
        self.canvas.create_line(600,166,850,166, fill='black', width=1)
        #2
        self.canvas.create_line(350,185,590,185, fill='black', width=1)
        self.canvas.create_line(350,195,590,195, fill='black', width=1)
        self.canvas.create_line(350,197,590,197, fill='black', width=1)
        self.canvas.create_line(350,206,590,206, fill='black', width=1)
        
        self.canvas.create_line(600,185,850,185, fill='black', width=1)
        self.canvas.create_line(600,195,850,195, fill='black', width=1)
        self.canvas.create_line(600,197,850,197, fill='black', width=1)
        self.canvas.create_line(600,206,850,206, fill='black', width=1)
        #3
        self.canvas.create_line(350,225,590,225, fill='black', width=1)
        self.canvas.create_line(350,235,590,235, fill='black', width=1)
        self.canvas.create_line(350,237,590,237, fill='black', width=1)
        self.canvas.create_line(350,246,590,246, fill='black', width=1)
        
        self.canvas.create_line(600,225,850,225, fill='black', width=1)
        self.canvas.create_line(600,235,850,235, fill='black', width=1)
        self.canvas.create_line(600,237,850,237, fill='black', width=1)
        self.canvas.create_line(600,246,850,246, fill='black', width=1)
        self.canvas.create_line(585,460,765,460, fill='black', width=3)
        self.canvas.create_line(365,460,550,460, fill='black', width=3)
        self.canvas.create_line(585,515,765,515, fill='black', width=3)
        self.canvas.create_line(365,515,550,515, fill='black', width=3)
        self.canvas.create_line(765,516,765, 460, fill='black', width=3)
        self.canvas.create_line(550,516,550, 460, fill='black', width=3)
        self.canvas.create_line(585,516,585, 460, fill='black', width=3)
        self.canvas.create_line(365,516,365, 460, fill='black', width=3)
        self.canvas.create_line(300,100,300,225, fill='black', width=3)
        self.canvas.create_line(200,100,300,100, fill='black', width=3)
        self.canvas.create_line(200,100,200,120, fill='black', width=3)
        self.canvas.create_line(200,120,170,120, fill='black', width=3)
        self.canvas.create_line(170,120,160,160, fill='black', width=4)
        self.canvas.create_line(160,160,160,210, fill='black', width=4)
        self.canvas.create_line(160,210,150,210, fill='black', width=3)
        self.canvas.create_line(150,210,150,411, fill='black', width=3)
        self.canvas.create_line(270,225,300,225, fill='black', width=3)
        self.canvas.create_line(270,225,270,253, fill='black', width=3)
        self.canvas.create_line(270,253,300,253, fill='black', width=3)
        self.canvas.create_line(260,430,300, 253, fill='black', width=3)
        self.canvas.create_line(117,410,150,410, fill='black', width=3)
        self.canvas.create_line(100,430,260,430, fill='black', width=3)
        self.canvas.create_line(175,390,240,390, fill='black', width=3)
        self.canvas.create_line(240,390,267, 260, fill='black', width=3)
        self.canvas.create_line(175,260,268,260, fill='black', width=3)
        self.canvas.create_line(175,260,175,390, fill='black', width=3)
        self.canvas.create_line(293,430,333, 283, fill='black', width=3)
        self.canvas.create_line(323,430,360, 310, fill='black', width=3)
        self.canvas.create_line(333,283,370,283, fill='black', width=3)
        self.canvas.create_line(293,430,323,430, fill='black', width=3)
        self.canvas.create_line(348,308,362,310, fill='black', width=3)
        self.canvas.create_line(350,293,365,295, fill='black', width=3)
        self.canvas.create_line(370,282,365, 296, fill='black', width=3)


            
        # Coloring
        self.canvas.create_line(0,560,850,560, fill='lightgray', width=20)
        self.canvas.create_line(390,415,850,415, fill='lightgray', width=26)
        self.canvas.create_line(200,27,900,27, fill='lightgray', width=23)
        self.canvas.create_line(350,105,850,105, fill='lightgray', width=8)
        self.canvas.create_line(824,569,824, 410, fill='lightgray', width=51)
        self.canvas.create_line(924,16,924,299, fill='lightgray', width=50)
        self.canvas.create_line(415,292,949,292, fill='lightgray', width=13)
        self.canvas.create_line(377,427,420, 286, fill='lightgray', width=10)
        self.canvas.create_line(390,427,438, 286, fill='lightgray', width=10)
        self.canvas.create_line(380,427,428, 286, fill='lightgray', width=10)
        #coloring the rtg box
        self.canvas.create_line(350,135,850,135, fill='orange', width=8)
        self.canvas.create_line(350,175,850,175, fill='orange', width=8)
        self.canvas.create_line(350,215,850,215, fill='orange', width=8)
        self.canvas.create_line(350,255,850,255, fill='orange', width=8)
        #1
        self.canvas.create_line(350,150,590,150, fill='lightgray', width=8)
        self.canvas.create_line(600,150,850,150, fill='lightgray', width=8)
        self.canvas.create_line(350,161,590,161, fill='lightgray', width=8)
        self.canvas.create_line(600,161,850,161, fill='lightgray', width=8)
        #2
        self.canvas.create_line(350,190,590,190, fill='lightgray', width=9)
        self.canvas.create_line(600,190,850,190, fill='lightgray', width=9)
        self.canvas.create_line(350,201,590,201, fill='lightgray', width=9)
        self.canvas.create_line(600,201,850,201, fill='lightgray', width=9)
        
        self.canvas.create_line(350,230,590,230, fill='lightgray', width=9)
        self.canvas.create_line(600,230,850,230, fill='lightgray', width=9)
    
        self.canvas.create_line(600,241,850,241, fill='lightgray', width=9)
        self.canvas.create_line(587,487,765,487, fill='gold', width=51)
        self.canvas.create_line(367,487,550,487, fill='gold', width=51)
        self.canvas.create_line(116,420,260,420, fill='lightgray', width=17)
        self.canvas.create_line(156,212,156,413, fill='lightgray', width=10)
        self.canvas.create_line(157,211,157,224, fill='lightgray', width=12)
        self.canvas.create_line(255,425,294, 254, fill='lightgray', width=10)
        self.canvas.create_line(202,163,299,163, fill='lightgray', width=122)
        self.canvas.create_line(202,172,172,172, fill='lightgray', width=103)
        self.canvas.create_line(174,122,164,160, fill='lightgray', width=6)
        self.canvas.create_line(170,145,170,224, fill='lightgray', width=15)
        self.canvas.create_line(160,407,260,407, fill='lightgray', width=30)
        self.canvas.create_line(250,400,279, 255, fill='lightgray', width=20)
        self.canvas.create_line(160,241,270,241, fill='lightgray', width=35)
        self.canvas.create_line(164,252,164,395, fill='lightgray', width=20)
        self.canvas.create_line(270,255,300,255, fill='lightgray', width=3)
        self.canvas.create_line(175,380,240,380, fill='gold', width=17)
        self.canvas.create_line(232,388,259, 262, fill='gold', width=15)
        self.canvas.create_line(175,268,266,268, fill='gold', width=15)
        self.canvas.create_line(180,268,180,380, fill='gold', width=10)
        self.canvas.create_line(212,268,212,380, fill='gold', width=59)
        self.canvas.create_line(245,268,245,310, fill='gold', width=10)
        self.canvas.create_line(185,227,185,258, fill='yellow', width=20)
        self.canvas.create_line(202,110,250,110, fill='lightgray', width=19)
        self.canvas.create_line(200,27,250,27, fill='lightgray', width=23)
        self.canvas.create_line(250,27,300,27, fill='lightgray', width=23)
        self.canvas.create_line(200,70,220,70, fill='black', width=19)
        self.canvas.create_line(220,70,230,70, fill='lightgray', width=15)
        self.canvas.create_line(180,505,230,505, fill='gray', width=80)
        self.canvas.create_line(160,475,250,475, fill='gray', width=10)
        self.canvas.create_line(160,495,250,495, fill='gray', width=10)
        self.canvas.create_line(160,515,250,515, fill='gray', width=10)
        self.canvas.create_line(160,535,250,535, fill='gray', width=10)
        self.canvas.create_line(165,475,175,475, fill='black', width=10) 
        self.canvas.create_line(165,495,175,495, fill='black', width=10)
        self.canvas.create_line(165,515,175,515, fill='black', width=10)
        self.canvas.create_line(165,535,175,535, fill='black', width=10)
        self.canvas.create_line(140,442,220,442, fill='red', width=19)# wb2
        self.canvas.create_line(282,390,300, 315, fill='red', width=21)# wb1
        self.canvas.create_line(302,370,306, 355, fill='black', width=10)
        self.canvas.create_line(302,429,342, 285, fill='lightgray', width=16)
        self.canvas.create_line(314,428,351, 310, fill='lightgray', width=16)
        self.canvas.create_line(334,288,367,288, fill='lightgray', width=7)
        self.canvas.create_line(331,330,342, 290, fill='yellow', width=17)
        self.canvas.create_line(220,215,270,215, fill='yellow', width=19)
        
        #unloading zone dots
        #part 1 
        #line 1
        self.canvas.create_line(380,467,382,467, fill='black', width=3)
        self.canvas.create_line(380,477,382,477, fill='black', width=3)
        self.canvas.create_line(380,487,382,487, fill='black', width=3)
        self.canvas.create_line(380,497,382,497, fill='black', width=3)
        self.canvas.create_line(380,507,382,507, fill='black', width=3)
       #line 2
        self.canvas.create_line(400,467,402,467, fill='black', width=3)
        self.canvas.create_line(400,477,402,477, fill='black', width=3)
        self.canvas.create_line(400,487,402,487, fill='black', width=3)
        self.canvas.create_line(400,497,402,497, fill='black', width=3)
        self.canvas.create_line(400,507,402,507, fill='black', width=3)
       #line 3
        self.canvas.create_line(420,467,422,467, fill='black', width=3)
        self.canvas.create_line(420,477,422,477, fill='black', width=3)
        self.canvas.create_line(420,487,422,487, fill='black', width=3)
        self.canvas.create_line(420,497,422,497, fill='black', width=3)
        self.canvas.create_line(420,507,422,507, fill='black', width=3)
      #line 4
        self.canvas.create_line(440,467,442,467, fill='black', width=3)
        self.canvas.create_line(440,477,442,477, fill='black', width=3)
        self.canvas.create_line(440,487,442,487, fill='black', width=3)
        self.canvas.create_line(440,497,442,497, fill='black', width=3)
        self.canvas.create_line(440,507,442,507, fill='black', width=3)
        #line 5
        self.canvas.create_line(460,467,462,467, fill='black', width=3)
        self.canvas.create_line(460,477,462,477, fill='black', width=3)
        self.canvas.create_line(460,487,462,487, fill='black', width=3)
        self.canvas.create_line(460,497,462,497, fill='black', width=3)
        self.canvas.create_line(460,507,462,507, fill='black', width=3)
        #line 6
        self.canvas.create_line(480,467,482,467, fill='black', width=3)
        self.canvas.create_line(480,477,482,477, fill='black', width=3)
        self.canvas.create_line(480,487,482,487, fill='black', width=3)
        self.canvas.create_line(480,497,482,497, fill='black', width=3)
        self.canvas.create_line(480,507,482,507, fill='black', width=3)
        #line 7
        self.canvas.create_line(500,467,502,467, fill='black', width=3)
        self.canvas.create_line(500,477,502,477, fill='black', width=3)
        self.canvas.create_line(500,487,502,487, fill='black', width=3)
        self.canvas.create_line(500,497,502,497, fill='black', width=3)
        self.canvas.create_line(500,507,502,507, fill='black', width=3)
        #line 8
        self.canvas.create_line(520,467,522,467, fill='black', width=3)
        self.canvas.create_line(520,477,522,477, fill='black', width=3)
        self.canvas.create_line(520,487,522,487, fill='black', width=3)
        self.canvas.create_line(520,497,522,497, fill='black', width=3)
        self.canvas.create_line(520,507,522,507, fill='black', width=3)
        #line 9
        self.canvas.create_line(540,467,542,467, fill='black', width=3)
        self.canvas.create_line(540,477,542,477, fill='black', width=3)
        self.canvas.create_line(540,487,542,487, fill='black', width=3)
        self.canvas.create_line(540,497,542,497, fill='black', width=3)
        self.canvas.create_line(540,507,542,507, fill='black', width=3)
        
        #part 2 
        #line 1
        self.canvas.create_line(600,467,602,467, fill='black', width=3)
        self.canvas.create_line(600,477,602,477, fill='black', width=3)
        self.canvas.create_line(600,487,602,487, fill='black', width=3)
        self.canvas.create_line(600,497,602,497, fill='black', width=3)
        self.canvas.create_line(600,507,602,507, fill='black', width=3)
        #line 2
        self.canvas.create_line(620,467,622,467, fill='black', width=3)
        self.canvas.create_line(620,477,622,477, fill='black', width=3)
        self.canvas.create_line(620,487,622,487, fill='black', width=3)
        self.canvas.create_line(620,497,622,497, fill='black', width=3)
        self.canvas.create_line(620,507,622,507, fill='black', width=3)
        #line 3
        self.canvas.create_line(640,467,642,467, fill='black', width=3)
        self.canvas.create_line(640,477,642,477, fill='black', width=3)
        self.canvas.create_line(640,487,642,487, fill='black', width=3)
        self.canvas.create_line(640,497,642,497, fill='black', width=3)
        self.canvas.create_line(640,507,642,507, fill='black', width=3)
        #line 4
        self.canvas.create_line(660,467,662,467, fill='black', width=3)
        self.canvas.create_line(660,477,662,477, fill='black', width=3)
        self.canvas.create_line(660,487,662,487, fill='black', width=3)
        self.canvas.create_line(660,497,662,497, fill='black', width=3)
        self.canvas.create_line(660,507,662,507, fill='black', width=3)
         #line 5
        self.canvas.create_line(680,467,682,467, fill='black', width=3)
        self.canvas.create_line(680,477,682,477, fill='black', width=3)
        self.canvas.create_line(680,487,682,487, fill='black', width=3)
        self.canvas.create_line(680,497,682,497, fill='black', width=3)
        self.canvas.create_line(680,507,682,507, fill='black', width=3)
         #line 6
        self.canvas.create_line(700,467,702,467, fill='black', width=3)
        self.canvas.create_line(700,477,702,477, fill='black', width=3)
        self.canvas.create_line(700,487,702,487, fill='black', width=3)
        self.canvas.create_line(700,497,702,497, fill='black', width=3)
        self.canvas.create_line(700,507,702,507, fill='black', width=3)
         #line 7
        self.canvas.create_line(720,467,722,467, fill='black', width=3)
        self.canvas.create_line(720,477,722,477, fill='black', width=3)
        self.canvas.create_line(720,487,722,487, fill='black', width=3)
        self.canvas.create_line(720,497,722,497, fill='black', width=3)
        self.canvas.create_line(720,507,722,507, fill='black', width=3)
         #line 8
        self.canvas.create_line(740,467,742,467, fill='black', width=3)
        self.canvas.create_line(740,477,742,477, fill='black', width=3)
        self.canvas.create_line(740,487,742,487, fill='black', width=3)
        self.canvas.create_line(740,497,742,497, fill='black', width=3)
        self.canvas.create_line(740,507,742,507, fill='black', width=3)
         #line 9
        self.canvas.create_line(757,467,759,467, fill='black', width=3)
        self.canvas.create_line(757,477,759,477, fill='black', width=3)
        self.canvas.create_line(757,487,759,487, fill='black', width=3)
        self.canvas.create_line(757,497,759,497, fill='black', width=3)
        self.canvas.create_line(757,507,759,507, fill='black', width=3)
        
        #rtg lines
        #1
        self.canvas.create_line(350,145,350, 155, fill='black', width=1)
        self.canvas.create_line(360,145,360, 155, fill='black', width=1)
        self.canvas.create_line(370,145,370, 155, fill='black', width=1)
        self.canvas.create_line(380,145,380, 155, fill='black', width=1)
        self.canvas.create_line(390,145,390, 155, fill='black', width=1)
        self.canvas.create_line(400,145,400, 155, fill='black', width=1)
        self.canvas.create_line(410,145,410, 155, fill='black', width=1)
        self.canvas.create_line(420,145,420, 155, fill='black', width=1)
        self.canvas.create_line(430,145,430, 155, fill='black', width=1)
        self.canvas.create_line(440,145,440, 155, fill='black', width=1)
        self.canvas.create_line(450,145,450, 155, fill='black', width=1)
        self.canvas.create_line(460,145,460, 155, fill='black', width=1)
        self.canvas.create_line(470,145,470, 155, fill='black', width=1)
        self.canvas.create_line(480,145,480, 155, fill='black', width=1)
        self.canvas.create_line(490,145,490, 155, fill='black', width=1)
        self.canvas.create_line(500,145,500, 155, fill='black', width=1)
        self.canvas.create_line(510,145,510, 155, fill='black', width=1)
        self.canvas.create_line(520,145,520, 155, fill='black', width=1)
        self.canvas.create_line(530,145,530, 155, fill='black', width=1)
        self.canvas.create_line(540,145,540, 155, fill='black', width=1)
        self.canvas.create_line(550,145,550, 155, fill='black', width=1)
        self.canvas.create_line(560,145,560, 155, fill='black', width=1)
        self.canvas.create_line(570,145,570, 155, fill='black', width=1)
        self.canvas.create_line(580,145,580, 155, fill='black', width=1)
        self.canvas.create_line(590,145,590, 155, fill='black', width=1)
        
        #2
        self.canvas.create_line(350,185,350, 195, fill='black', width=1)
        self.canvas.create_line(370,185,370, 195, fill='black', width=1)
        self.canvas.create_line(380,185,380, 195, fill='black', width=1)
        self.canvas.create_line(390,185,390, 195, fill='black', width=1)
        self.canvas.create_line(400,185,400, 195, fill='black', width=1)
        self.canvas.create_line(410,185,410, 195, fill='black', width=1)
        self.canvas.create_line(420,185,420, 195, fill='black', width=1)
        self.canvas.create_line(430,185,430, 195, fill='black', width=1)
        self.canvas.create_line(440,185,440, 195, fill='black', width=1)
        self.canvas.create_line(450,185,450, 195, fill='black', width=1)
        self.canvas.create_line(460,185,460, 195, fill='black', width=1)
        self.canvas.create_line(470,185,470, 195, fill='black', width=1)
        self.canvas.create_line(480,185,480, 195, fill='black', width=1)
        self.canvas.create_line(490,185,490, 195, fill='black', width=1)
        self.canvas.create_line(500,185,500, 195, fill='black', width=1)
        self.canvas.create_line(510,185,510, 195, fill='black', width=1)
        self.canvas.create_line(520,185,520, 195, fill='black', width=1)
        self.canvas.create_line(530,185,530, 195, fill='black', width=1)
        self.canvas.create_line(540,185,540, 195, fill='black', width=1)
        self.canvas.create_line(550,185,550, 195, fill='black', width=1)
        self.canvas.create_line(560,185, 560,195, fill='black', width=1)
        self.canvas.create_line(570,185,570, 195, fill='black', width=1)
        self.canvas.create_line(580,185,580, 195, fill='black', width=1)
        self.canvas.create_line(590,185,590, 195, fill='black', width=1)
        #3
        self.canvas.create_line(350,157,350, 166, fill='black', width=1)
        self.canvas.create_line(360,157,360, 166, fill='black', width=1)
        self.canvas.create_line(370,157,370, 166, fill='black', width=1)
        self.canvas.create_line(380,157,380, 166, fill='black', width=1)
        self.canvas.create_line(390,157,390, 166, fill='black', width=1)
        self.canvas.create_line(400,157,400, 166, fill='black', width=1)
        self.canvas.create_line(410,157,410, 166, fill='black', width=1)
        self.canvas.create_line(420,157,420, 166, fill='black', width=1)
        self.canvas.create_line(430,157,430, 166, fill='black', width=1)
        self.canvas.create_line(440,157,440, 166, fill='black', width=1)
        self.canvas.create_line(450,157,450, 166, fill='black', width=1)
        self.canvas.create_line(460,157,460, 166, fill='black', width=1)
        self.canvas.create_line(470,157,470, 166, fill='black', width=1)
        self.canvas.create_line(480,157,480, 166, fill='black', width=1)
        self.canvas.create_line(490,157,490, 166, fill='black', width=1)
        self.canvas.create_line(500,157,500, 166, fill='black', width=1)
        self.canvas.create_line(510,157,510, 166, fill='black', width=1)
        self.canvas.create_line(520,157,520, 166, fill='black', width=1)
        self.canvas.create_line(530,157,530, 166, fill='black', width=1)
        self.canvas.create_line(540,157,540, 166, fill='black', width=1)
        self.canvas.create_line(550,157,550, 166, fill='black', width=1)
        self.canvas.create_line(560,157,560, 166, fill='black', width=1)
        self.canvas.create_line(570,157,570, 166, fill='black', width=1)
        self.canvas.create_line(580,157,580, 166, fill='black', width=1)
        self.canvas.create_line(590,157,590, 166, fill='black', width=1)
        #4
        self.canvas.create_line(350,197,350, 206, fill='black', width=1)
        self.canvas.create_line(360,197,360, 206, fill='black', width=1)
        self.canvas.create_line(370,197,370, 206, fill='black', width=1)
        self.canvas.create_line(380,197,380, 206, fill='black', width=1)
        self.canvas.create_line(390,197,390, 206, fill='black', width=1)
        self.canvas.create_line(400,197,400, 206, fill='black', width=1)
        self.canvas.create_line(410,197,410, 206, fill='black', width=1)
        self.canvas.create_line(420,197,420, 206, fill='black', width=1)
        self.canvas.create_line(430,197,430, 206, fill='black', width=1)
        self.canvas.create_line(440,197,440, 206, fill='black', width=1)
        self.canvas.create_line(450,197,450, 206, fill='black', width=1)
        self.canvas.create_line(460,197,460, 206, fill='black', width=1)
        self.canvas.create_line(470,197,470, 206, fill='black', width=1)
        self.canvas.create_line(480,197,480, 206, fill='black', width=1)
        self.canvas.create_line(490,197,490, 206, fill='black', width=1)
        self.canvas.create_line(500,197,500, 206, fill='black', width=1)
        self.canvas.create_line(510,197,510, 206, fill='black', width=1)
        self.canvas.create_line(520,197,520, 206, fill='black', width=1)
        self.canvas.create_line(530,197,530, 206, fill='black', width=1)
        self.canvas.create_line(540,197,540, 206, fill='black', width=1)
        self.canvas.create_line(550,197,550, 206, fill='black', width=1)
        self.canvas.create_line(560,197, 560,206, fill='black', width=1)
        self.canvas.create_line(570,197,570, 206, fill='black', width=1)
        self.canvas.create_line(580,197,580, 206, fill='black', width=1)
        self.canvas.create_line(590,197,590, 206, fill='black', width=1)
        #5
        self.canvas.create_line(350,226,350, 236, fill='black', width=1)
        self.canvas.create_line(360,226,360, 236, fill='black', width=1)
        self.canvas.create_line(370,226,370, 236, fill='black', width=1)
        self.canvas.create_line(380,226,380, 236, fill='black', width=1)
        self.canvas.create_line(390,226,390, 236, fill='black', width=1)
        self.canvas.create_line(400,226,400, 236, fill='black', width=1)
        self.canvas.create_line(410,226,410, 236, fill='black', width=1)
        self.canvas.create_line(420,226,420, 236, fill='black', width=1)
        self.canvas.create_line(430,226,430, 236, fill='black', width=1)
        self.canvas.create_line(440,226,440, 236, fill='black', width=1)
        self.canvas.create_line(450,226,450, 236, fill='black', width=1)
        self.canvas.create_line(460,226,460, 236, fill='black', width=1)
        self.canvas.create_line(470,226,470, 236, fill='black', width=1)
        self.canvas.create_line(480,226,480, 236, fill='black', width=1)
        self.canvas.create_line(490,226,490, 236, fill='black', width=1)
        self.canvas.create_line(500,226,500, 236, fill='black', width=1)
        self.canvas.create_line(510,226,510, 236, fill='black', width=1)
        self.canvas.create_line(520,226,520, 236, fill='black', width=1)
        self.canvas.create_line(530,226,530, 236, fill='black', width=1)
        self.canvas.create_line(540,226,540, 236, fill='black', width=1)
        self.canvas.create_line(550,226,550, 236, fill='black', width=1)
        self.canvas.create_line(560,226, 560,236, fill='black', width=1)
        self.canvas.create_line(570,226,570, 236, fill='black', width=1)
        self.canvas.create_line(580,226,580, 236, fill='black', width=1)
        self.canvas.create_line(590,226,590, 236, fill='black', width=1)
        #6
        self.canvas.create_line(350,237,350, 246, fill='black', width=1)
        self.canvas.create_line(590,237,590, 246, fill='black', width=1)
        
        #1
        self.canvas.create_line(600,145,600, 155, fill='black', width=1)
        self.canvas.create_line(610,145,610, 155, fill='black', width=1)
        self.canvas.create_line(620,145,620, 155, fill='black', width=1)
        self.canvas.create_line(630,145,630, 155, fill='black', width=1)
        self.canvas.create_line(640,145,640, 155, fill='black', width=1)
        self.canvas.create_line(650,145,650, 155, fill='black', width=1)
        self.canvas.create_line(660,145,660, 155, fill='black', width=1)
        self.canvas.create_line(670,145,670, 155, fill='black', width=1)
        self.canvas.create_line(680,145,680, 155, fill='black', width=1)
        self.canvas.create_line(690,145,690, 155, fill='black', width=1)
        self.canvas.create_line(700,145,700, 155, fill='black', width=1)
        self.canvas.create_line(710,145,710, 155, fill='black', width=1)
        self.canvas.create_line(720,145,720, 155, fill='black', width=1)
        self.canvas.create_line(730,145,730, 155, fill='black', width=1)
        self.canvas.create_line(740,145,740, 155, fill='black', width=1)
        self.canvas.create_line(750,145,750, 155, fill='black', width=1)
        self.canvas.create_line(760,145,760, 155, fill='black', width=1)
        self.canvas.create_line(770,145,770, 155, fill='black', width=1)
        self.canvas.create_line(780,145,780, 155, fill='black', width=1)
        self.canvas.create_line(790,145,790, 155, fill='black', width=1)
        self.canvas.create_line(800,145,800, 155, fill='black', width=1)
        self.canvas.create_line(810,145,810, 155, fill='black', width=1)
        self.canvas.create_line(820,145,820, 155, fill='black', width=1)
        self.canvas.create_line(830,145,830, 155, fill='black', width=1)
        self.canvas.create_line(840,145,840, 155, fill='black', width=1)
        self.canvas.create_line(850,145,850, 155, fill='black', width=1)
        #2
        self.canvas.create_line(600,157,600, 166, fill='black', width=1)
        self.canvas.create_line(610,157,610, 166, fill='black', width=1)
        self.canvas.create_line(620,157,620, 166, fill='black', width=1)
        self.canvas.create_line(630,157,630, 166, fill='black', width=1)
        self.canvas.create_line(640,157,640, 166, fill='black', width=1)
        self.canvas.create_line(650,157,650, 166, fill='black', width=1)
        self.canvas.create_line(660,157,660, 166, fill='black', width=1)
        self.canvas.create_line(670,157,670, 166, fill='black', width=1)
        self.canvas.create_line(680,157,680, 166, fill='black', width=1)
        self.canvas.create_line(690,157,690, 166, fill='black', width=1)
        self.canvas.create_line(700,157,700, 166, fill='black', width=1)
        self.canvas.create_line(710,157,710, 166, fill='black', width=1)
        self.canvas.create_line(720,157,720, 166, fill='black', width=1)
        self.canvas.create_line(730,157,730, 166, fill='black', width=1)
        self.canvas.create_line(740,157,740, 166, fill='black', width=1)
        self.canvas.create_line(750,157,750, 166, fill='black', width=1)
        self.canvas.create_line(760,157,760, 166, fill='black', width=1)
        self.canvas.create_line(770,157,770, 166, fill='black', width=1)
        self.canvas.create_line(780,157,780, 166, fill='black', width=1)
        self.canvas.create_line(790,157,790, 166, fill='black', width=1)
        self.canvas.create_line(800,157,800, 166, fill='black', width=1)
        self.canvas.create_line(810,157,810, 166, fill='black', width=1)
        self.canvas.create_line(820,157,820, 166, fill='black', width=1)
        self.canvas.create_line(830,157,830, 166, fill='black', width=1)
        self.canvas.create_line(840,157,840, 166, fill='black', width=1)
        self.canvas.create_line(850,157,850, 166, fill='black', width=1)
        #3
        self.canvas.create_line(600,185,600, 195, fill='black', width=1)
        self.canvas.create_line(610,185,610, 195, fill='black', width=1)
        self.canvas.create_line(620,185,620, 195, fill='black', width=1)
        self.canvas.create_line(630,185,630, 195, fill='black', width=1)
        self.canvas.create_line(640,185,640, 195, fill='black', width=1)
        self.canvas.create_line(650,185,650, 195, fill='black', width=1)
        self.canvas.create_line(660,185,660, 195, fill='black', width=1)
        self.canvas.create_line(670,185,670, 195, fill='black', width=1)
        self.canvas.create_line(680,185,680, 195, fill='black', width=1)
        self.canvas.create_line(690,185,690, 195, fill='black', width=1)
        self.canvas.create_line(700,185,700, 195, fill='black', width=1)
        self.canvas.create_line(710,185,710, 195, fill='black', width=1)
        self.canvas.create_line(720,185,720, 195, fill='black', width=1)
        self.canvas.create_line(730,185,730, 195, fill='black', width=1)
        self.canvas.create_line(740,185,740, 195, fill='black', width=1)
        self.canvas.create_line(750,185,750, 195, fill='black', width=1)
        self.canvas.create_line(760,185,760, 195, fill='black', width=1)
        self.canvas.create_line(770,185,770, 195, fill='black', width=1)
        self.canvas.create_line(780,185,780, 195, fill='black', width=1)
        self.canvas.create_line(790,185,790, 195, fill='black', width=1)
        self.canvas.create_line(800,185,800, 195, fill='black', width=1)
        self.canvas.create_line(810,185,810, 195, fill='black', width=1)
        self.canvas.create_line(820,185,820, 195, fill='black', width=1)
        self.canvas.create_line(830,185,830, 195, fill='black', width=1)
        self.canvas.create_line(840,185,840, 195, fill='black', width=1)
        self.canvas.create_line(850,185,850, 195, fill='black', width=1)
        #4
        self.canvas.create_line(600,197,600, 206, fill='black', width=1)
        self.canvas.create_line(610,197,610, 206, fill='black', width=1)
        self.canvas.create_line(620,197,620, 206, fill='black', width=1)
        self.canvas.create_line(630,197,630, 206, fill='black', width=1)
        self.canvas.create_line(640,197,640, 206, fill='black', width=1)
        self.canvas.create_line(650,197,650, 206, fill='black', width=1)
        self.canvas.create_line(660,197,660, 206, fill='black', width=1)
        self.canvas.create_line(670,197,670, 206, fill='black', width=1)
        self.canvas.create_line(690,197,690, 206, fill='black', width=1)
        self.canvas.create_line(700,197,700, 206, fill='black', width=1)
        self.canvas.create_line(710,197,710, 206, fill='black', width=1)
        self.canvas.create_line(720,197,720, 206, fill='black', width=1)
        self.canvas.create_line(730,197,730, 206, fill='black', width=1)
        self.canvas.create_line(740,197,740, 206, fill='black', width=1)
        self.canvas.create_line(750,197,750, 206, fill='black', width=1)
        self.canvas.create_line(760,197,760, 206, fill='black', width=1)
        self.canvas.create_line(770,197,770, 206, fill='black', width=1)
        self.canvas.create_line(780,197,780, 206, fill='black', width=1)
        self.canvas.create_line(790,197,790, 206, fill='black', width=1)
        self.canvas.create_line(800,197,800, 206, fill='black', width=1)
        self.canvas.create_line(810,197,810, 206, fill='black', width=1)
        self.canvas.create_line(820,197,820, 206, fill='black', width=1)
        self.canvas.create_line(830,197,830, 206, fill='black', width=1)
        self.canvas.create_line(840,197,840, 206, fill='black', width=1)
        self.canvas.create_line(850,197,850, 206, fill='black', width=1)
        #5
        self.canvas.create_line(600,226,600, 236, fill='black', width=1)
        self.canvas.create_line(610,226,610, 236, fill='black', width=1)
        self.canvas.create_line(620,226,620, 236, fill='black', width=1)
        self.canvas.create_line(630,226,630, 236, fill='black', width=1)
        self.canvas.create_line(640,226,640, 236, fill='black', width=1)
        self.canvas.create_line(650,226,650, 236, fill='black', width=1)
        self.canvas.create_line(660,226,660, 236, fill='black', width=1)
        self.canvas.create_line(670,226,670, 236, fill='black', width=1)
        self.canvas.create_line(680,226,680, 236, fill='black', width=1)
        self.canvas.create_line(690,226,690, 236, fill='black', width=1)
        self.canvas.create_line(700,226,700, 236, fill='black', width=1)
        self.canvas.create_line(710,226,710, 236, fill='black', width=1)
        self.canvas.create_line(720,226,720, 236, fill='black', width=1)
        self.canvas.create_line(730,226,730, 236, fill='black', width=1)
        self.canvas.create_line(740,226,740, 236, fill='black', width=1)
        self.canvas.create_line(750,226,750, 236, fill='black', width=1)
        self.canvas.create_line(760,226,760, 236, fill='black', width=1)
        self.canvas.create_line(770,226,770, 236, fill='black', width=1)
        self.canvas.create_line(780,226,780, 236, fill='black', width=1)
        self.canvas.create_line(790,226,790, 236, fill='black', width=1)
        self.canvas.create_line(800,226,800, 236, fill='black', width=1)
        self.canvas.create_line(810,226,810, 236, fill='black', width=1)
        self.canvas.create_line(820,226,820, 236, fill='black', width=1)
        self.canvas.create_line(830,226,830, 236, fill='black', width=1)
        self.canvas.create_line(840,226,840, 236, fill='black', width=1)
        self.canvas.create_line(850,226,850, 236, fill='black', width=1)
        #7
        self.canvas.create_line(600,237,600, 246, fill='black', width=1)
        self.canvas.create_line(610,237,610, 246, fill='black', width=1)
        self.canvas.create_line(620,237,620, 246, fill='black', width=1)
        self.canvas.create_line(630,237,630, 246, fill='black', width=1)
        self.canvas.create_line(640,237,640, 246, fill='black', width=1)
        self.canvas.create_line(650,237,650, 246, fill='black', width=1)
        self.canvas.create_line(660,237,660, 246, fill='black', width=1)
        self.canvas.create_line(670,237,670, 246, fill='black', width=1)
        self.canvas.create_line(680,237,680, 246, fill='black', width=1)
        self.canvas.create_line(690,237,690, 246, fill='black', width=1)
        self.canvas.create_line(700,237,700, 246, fill='black', width=1)
        self.canvas.create_line(710,237,710, 246, fill='black', width=1)
        self.canvas.create_line(720,237,720, 246, fill='black', width=1)
        self.canvas.create_line(730,237,730, 246, fill='black', width=1)
        self.canvas.create_line(740,237,740, 246, fill='black', width=1)
        self.canvas.create_line(750,237,750, 246, fill='black', width=1)
        self.canvas.create_line(760,237,760, 246, fill='black', width=1)
        self.canvas.create_line(770,237,770, 246, fill='black', width=1)
        self.canvas.create_line(780,237,780, 246, fill='black', width=1)
        self.canvas.create_line(790,237,790, 246, fill='black', width=1)
        self.canvas.create_line(800,237,800, 246, fill='black', width=1)
        self.canvas.create_line(810,237,810, 246, fill='black', width=1)
        self.canvas.create_line(820,237,820, 246, fill='black', width=1)
        self.canvas.create_line(830,237,830, 246, fill='black', width=1)
        self.canvas.create_line(840,237,840, 246, fill='black', width=1)
        self.canvas.create_line(850,237,850, 246, fill='black', width=1)
    
    
        
        
        
        
        # Add text diagonally without animation
        text_id = self.canvas.create_text(412, 350, text="QUAY 3", font=("Arial", 10, "bold"), fill="black", angle=72)
        text_id = self.canvas.create_text(941, 165, text="QUAY 1", font=("Arial", 10, "bold"), fill="black", angle=90)
        text_id = self.canvas.create_text(293, 353, text="WB1", font=("Arial", 10, "bold"), fill="black", angle=75)
        text_id = self.canvas.create_text(610, 293, text="QUAY 2", font=("Arial", 10, "bold"), fill="black", angle=0)
        text_id = self.canvas.create_text(600, 414, text="QUAY 4", font=("Arial", 10, "bold"), fill="black", angle=0)
        text_id = self.canvas.create_text(675, 487, text="Unloading Zone", font=("Arial", 10, "bold"), fill="black", angle=0)
        text_id = self.canvas.create_text(457, 487, text="Unloading Zone ", font=("Arial", 10, "bold"), fill="black", angle=0)
        text_id = self.canvas.create_text(860, 190, text="RTG zone", font=("Arial", 10, "bold"), fill="black", angle=90)
        text_id = self.canvas.create_text(900, 350, text="SEA", font=("Arial", 10, "bold"), fill="black", angle=0)
        text_id = self.canvas.create_text(205, 500, text="Gate", font=("Arial", 10, "bold"), fill="black", angle=0)
        text_id = self.canvas.create_text(205, 515, text="In", font=("Arial", 10, "bold"), fill="black", angle=0)
        text_id = self.canvas.create_text(180, 70, text="Gate", font=("Arial", 10, "bold"), fill="black", angle=0)
        text_id = self.canvas.create_text(180, 85, text="Out", font=("Arial", 10, "bold"), fill="black", angle=0)
        
        text_id = self.canvas.create_text(180,442 , text="WB2", font=("Arial", 10, "bold"), fill="black", angle=0)
        
            
        

    def add_text(self, x:int, y:int, text:str):
        self.canvas.create_text(x, y, text=text, font=("Arial", 14), fill='black')
        
# Create the main Tkinter window
root = tk.Tk()
# Create an instance of the ThreeLinesBoxApp class
app = Queue(root)
# Run the Tkinter event loop
root.mainloop()
```
